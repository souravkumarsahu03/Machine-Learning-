# Decision Boundary/ Surfaces

* It is only applicable for classification.
* And it will work on 2_Dimensional data.
* That is 2 input features only and 1 output feature.

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_redundant=0, random_state=23)
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X,y)
```

```python
import seaborn as sns
sns.scatterplot(x = X[:,0], y= X[:,1], hue= y)
```

```python
!pip install mlxtend
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X, y=y, clf=knn)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X,y)
```

```python
plot_decision_regions(X=X, y=y, clf=gnb)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X,y)

plot_decision_regions(X=X, y=y, clf=lr)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X,y)

plot_decision_regions(X=X, y=y, clf=dt)
```

# Finding Optimal K Value

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,y)

plot_decision_regions(X=X, y=y, clf=lr)
```
