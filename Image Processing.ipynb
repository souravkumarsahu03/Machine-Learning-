# Combining the Images

```python
# gray = cv2.cvtColor(virat,cv2.COLOR_BGR2GRAY)
hls = cv2.cvtColor(virat,cv2.COLOR_BGR2HLS)
hsv = cv2.cvtColor(virat,cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(virat,cv2.COLOR_BGR2LAB)
bgr = cv2.cvtColor(virat,cv2.COLOR_BGR2RGB)
```

```python
import numpy as np
```

```python
img1 = np.hstack([virat,hsv])
img2 = np.hstack([lab,bgr])
img = np.vstack([img1,img2])
```

```python
ressized = cv2.resize(img, (800,500))
cv2.imshow('Virat',ressized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
