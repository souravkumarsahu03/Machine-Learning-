### Accessing webcam

```python
import cv2

video = cv2.VideoCapture(0)   #For webcam

while True:
    sucess, Frame = video.read()

    if not sucess:
        break

    cv2.imshow('My Face', Frame)

    if cv2.waitKey(1) & 255 == ord('q'):
        video.release()
        cv2.destroyAllWindows()
```

```python
import cv2

video = cv2.VideoCapture(0)   #For webcam
video.set(220,120)

while True:
    sucess, Frame = video.read()

    if not sucess:
        break

    cv2.putText(img = Frame,
               text= 'Face Detected',
                org = (120, 90),
               fontFace= cv2.FONT_HERSHEY_COMPLEX,
               fontScale= 1,
               thickness = 2,
               color= (255,255,255))

    cv2.rectangle(img = Frame, pt1 = (200,100), pt2 = (500,300), color= (0,255,10), thickness= 2)
    cv2.imshow('My Face', cv2.cvtColor(Frame, cv2.COLOR_BGR2GRAY))

    if cv2.waitKey(1) & 255 == ord('q'):
        video.release()
        cv2.destroyAllWindows()
```
