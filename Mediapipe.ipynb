```python
import cv2
import mediapipe as mp
import warnings as wr
wr.filterwarnings('ignore')
```

```python
mp.__version__
```

```python
face_detection = mp.solutions.face_detection.FaceDetection(min_detection_confidence= 0.7, model_selection= 0)
video = cv2.VideoCapture(0)

while True:
    sucess, Frames = video.read()
    rgb_image = cv2.cvtColor(Frames, cv2.COLOR_BGR2RGB)
    results = face_detection.process(rgb_image)

    if results.detections:
        for detections in results.detections:
            bbox = detections.location_data.relative_bounding_box
            x_min = int(bbox.xmin * Frames.shape[1])
            y_min = int(bbox.ymin*Frames.shape[0])
            x_max = int(bbox.width * Frames.shape[1])
            y_max = int(bbox.height * Frames.shape[0])

    score = round(detections.score[0], 2) * 100

    cv2.rectangle(img = Frames, pt1 = (x_min, y_min), pt2 = (x_min + x_max,y_min +  y_max), thickness = 2, color= (0,255,0))
    cv2.putText(img = Frames,
               text = f'Face Detected: {score}%',
               org = (x_min, y_min),
               fontScale = 1,
               thickness = 1,
               fontFace = cv2.FONT_HERSHEY_COMPLEX, 
               color = (0,255,0))
    cv2.imshow('Face Detection Model', Frames)

    if cv2.waitKey(2) & 255 == ord('q'):
        break
 
video.release()
cv2.destroyAllWindows()
```
