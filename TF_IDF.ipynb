```python
text = ["i am data analyst",
       "i am data analyst,data scientist and data engineer",
       "i am ML engineer and AI engineer"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
```

```python
tf = TfidfVectorizer(norm='l1')
pd.DataFrame(tf.fit_transform(text).toarray(), columns=tf.get_feature_names_out())
```
