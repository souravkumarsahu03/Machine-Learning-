# Vector Operations

* Which deals with the 1-Dimensional data
* eg: [10,20,30]
* Types: Hadamard Product, Dot Product

# Hadamard Product

* Element wise multiplication.
* Used only if the arrays are same.
* Example column size of 1st array = column size of 2nd array

```python
import numpy as np
arr = np.array([[10,20],[90,30]])
arr2 = np.array([[30,20],[3,4]])
```

```python
arr* arr2
```

# Dot Product

* Only applicable when the size of column of 1st array is equals to the row size of the second array
* Can not be applied for the different dimensions.
* It is a topic of arraybroadcasting.

```python
arr1 = np.array([[10,20,30]])
arr2 = np.array([[1,2,3],[4,5,6]])
```

```python
np.dot(arr1, arr2)
```

As we can see we are getting array because of the array shape

```python
arr1 = np.array([[10,20,30]])
arr2 = np.array([[1,2,3],
                 [4,5,6],
                [8,9,2]])
```

```python
np.dot(arr1, arr2)
```

# Finding Optimal K-Value

To getting the optimal k-value we can use the following method:
* Heuristic Approach: k = √n
* K-Fold Cross Validation
* Hyperparameter Tuning
* Visualization methods

In general we are not using the heuristic approach instead we use somthing called Cross Validation

### K-Fold 

```python
from sklearn.datasets import make_classification  #This is used to get the preprocessed datasets to work with 
```

```python
X,y = make_classification(n_samples=10000, n_features=2, n_redundant=0, random_state=23)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=23, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=23)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
k_value = range(2,20)
accuracy = []
for i in k_value:
    knn = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn,X,y, scoring='accuracy', cv=kf).mean()
    accuracy.append(score)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=k_value, y= accuracy)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'X': k_value, 'Accuracy': accuracy})
```
