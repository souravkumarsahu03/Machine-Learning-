# Discritization/Binning

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['sepal_length'].min(),df['sepal_length'].max()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import KBinsDiscretizer
kbin = KBinsDiscretizer(n_bins=3, encode= 'ordinal')
kbin.fit_transform(df[['sepal_length']])
```

```python
kbin.bin_edges_
```

```python
from sklearn.preprocessing import KBinsDiscretizer
kbin = KBinsDiscretizer(n_bins=5, encode= 'ordinal', strategy='quantile')
kbin.fit_transform(df[['sepal_length']])
```

```python
kbin.bin_edges_
```

```python
from sklearn.preprocessing import KBinsDiscretizer
kbin = KBinsDiscretizer(n_bins=5, encode= 'ordinal', strategy='uniform')
kbin.fit_transform(df[['sepal_length']])
```

```python
kbin.bin_edges_
```
